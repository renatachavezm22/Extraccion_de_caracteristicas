### **Limpieza de datos**

```python
Micro_Retailer.isnull().sum()
```

```python
Micro_RetailerSN=Micro_Retailer.copy()
Micro_RetailerSN
```

```python
Micro_RetailerSN=Micro_RetailerSN.fillna(round(Micro_Retailer.mean(),1))
Micro_RetailerSN=Micro_RetailerSN.fillna(method='ffill')
Micro_RetailerSN
```

```python
Micro_RetailerSN.isnull().sum()
```

### **Rango Intercuartílico**

```python
fig = plt.figure(figsize =(15, 8))
Micro_RetailerSN.plot(kind='box', vert=False)
plt.title("Distribución de Datos en el DataFrame")
plt.show() 
```

```python
y=Micro_RetailerSN

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido\n", Limite_Superior_iqr)
print()
print("Limite inferior permitido\n", Limite_Inferior_iqr)
```

```python
outliers_iqr= Micro_RetailerSN[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
Micro_Retailer_clean= Micro_RetailerSN[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
Micro_Retailer_clean
```

### **Division de valores cuantitativos y cualitativos**

```python
Micro_Retailer_str = Micro_Retailer_clean.select_dtypes(include=['object'])
Micro_Retailer_str.columns
```

```python
Micro_Retailer_float = Micro_Retailer_clean.select_dtypes(include=['float'])
Micro_Retailer_float.columns
```

### **Verificación de limpieza**

```python
Micro_Retailer_float.isnull().sum()
```

```python
mr_float=Micro_Retailer_float.copy()
mr_float=mr_float.fillna(round(Micro_Retailer_float.mean(),1))
mr_float=mr_float.fillna(method='ffill')
```

```python
fig = plt.figure(figsize =(15, 8))
mr_float.plot(kind='box', vert=False)
plt.title('Nueva distribución de valores del DataFrame')
plt.show() #dibujamos el diagrama
```
